```python
# numPy    OLS     
from sklearn.datasets import make_regression
import statsmodels.api as sm

bias = 100
X0, y, w = make_regression( n_samples=200, n_features=1, bias=bias, noise=10, coef=True, random_state=1)
#statsmodels     add_constant 
X = sm.add_constant(X0)
y = y.reshape(len(y), 1)
print(y)
print(w) #x y  y=100+86.44794301x+
```

```python
                  
#OLS        
w = np.linalg.inv(X.T @ X) @ X.T @ y
w                   #   y^=99.79150869+86.96171201x
```

```python
#         
#   
x_new = np.linspace(np.min(X0), np.max(X0), 10)
X_new = sm.add_constant(x_new) #  
y_new = np.dot(X_new, w)

plt.scatter(X0, y, label=" ")
plt.plot(x_new, y_new, 'rs-', label=" ")
plt.xlabel("x")
plt.ylabel("y")
plt.title("  ")
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
#LinearRegression(fit_intercept=True) fit_intercept      
model = LinearRegression().fit(X0, y) # #   (   )
#coef_ :    , intercept_ :  
print(model.intercept_, model.coef_)
#      
model.predict([[-2], [-1], [0], [1], [2]])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("./data/datas/auto-mpg.csv", header=None)
#  
df.columns =['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'name']

print(df.info())  #   

print(df.describe())  #     
```

```python
#horsepower   
print(df['horsepower'].unique())
```

```python
# ? np.nan 
df['horsepower'].replace('?', np.nan, inplace=True)
```

```python
#   
df.dropna(subset=['horsepower'], axis=0, inplace=True)
```

```python
#   
df['horsepower'] = df['horsepower'].astype('float')
df.describe() #     
```

```python
#   (, )  - , , , 
ndf = df[['mpg', 'cylinders', 'horsepower', 'weight']]

#   Y          
ndf.plot(kind='scatter', x='weight', y='mpg', c='coral', s=10, figsize=(10, 5))
plt.show()
plt.close()
```

```python
#seaborn  
fig = plt.figure(figsize=(10,2))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

#seaborn  
sns.regplot(x='weight', y='mpg', data=ndf, ax=ax1)
sns.regplot(x='weight', y='mpg', data=ndf, ax=ax2, fit_reg=False)
plt.show()
plt.close()
```

```python
# seaborn   ( )
sns.jointplot(x='weight', y='mpg', data=ndf)
sns.jointplot(x='weight', y='mpg', kind='reg', data=ndf)  #  
plt.show()
plt.close()
```

```python
# seaborn       
sns.pairplot(ndf)
plt.show()
plt.close()
```

```python
#   train test ( mpg,  weight)
X=ndf[['weight']]
y=ndf['mpg']

from sklearn.model_selection import train_test_split
#test data set  30%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)

# 
print('train  data  : ', len(X_train))
print('test data  : ',  len(X_test))
```

```python
#train       
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)

#  
r_square = lr.score(X_test, y_test)
print(' :',r_square)

# ()
print(' :', lr.coef_)

# 
print(' :', lr.intercept_ )

#  
y_hat = lr.predict(X)

plt.figure(figsize=(10,5))
ax1 = sns.distplot(y, hist=False, label='y')
ax2 = sns.distplot(y_hat, hist=False, label='y_hat', ax=ax1)  # ax=ax1 :  
plt.show()
plt.close()
```
