```python
## Sequence Type
## list
## 임의의 객체를 순차적으로 저장한ㄴ 자료구조
## 기호로는 [] (대괄호를 이용)
myList = list()        # 비어있는 list를 생성 아무것도 가지지말고 리스트만들어
# 벡터와 다른점 : 벡터는 같은 데이터타입의 값만 들어와, 리스트는 다양하게
myList = []           # list를 literal로 표현
myList = [1,2,3]      # 3개의 값을 가지는 list를 생성
myList = [1,3.14,True,"아우성!"]    # 모든 데이터 타입
myList = [10,["Show","me","the","money"],3.14,True]     # 리스트를 중첩해서 가능
                                            # List안에 또 다른 자료구조들을 넣을 수 있어요
print(myList[2])
print(myList[1][3])
print(myList[-1])
print(myList[0:1])
# slicing : 원본에 준하여 뽑아냄
# slicing결과는 원본자료구조와 동일
print(myList[1][2:])
```

```python
## list의 연산

a = [1,2,3]
b = [4,5,6]
print(a+b)      # 연결함
print(a*3)      # 3번 반복
```

```python
## list의 indexing & slicing

a = [1,2,3]
a[0] = 5        # 5, 2, 3
a[0:1] = [7,8,9]   # 7, 8, 9, 2, 3  
a[1:] = []     # [1]
# list slicing한 부분을 다른 리스트로 대치
a[0] = [7,8,9]   # [[7,8,9],2,3]

# indexing : 그 위치에 있는 값을 알아내거나 뽑아내는거 ( 위치를 지정 )
# slicing : 그 자료형을 대치하거나 뽑아내는거 ( 영역을 지정 )

a = [1,2,3]
del a[1]      # [1,3]
```

# Sequence Type - tuple

```python
# Seuence Type : tuple
# list와 거의 동일
# 순차적인 구조
# 표현하는 방식이 달라요
# list는 대괄호로 표현 [1,2,3]
# tuple은 소괄호로 표현 (1,2,3)
a = ()     # empty tuple
print(type(a))
myList = [1]    # 요소가 1개인 list
print(type(myList))
myTuple = (1)   # 요소가 1개인 tuple
print(type(myTuple))
# tuple의 괄호가아니라 연산의 괄호로 인식
# list의 괄호로 표현하면 안돼
myTuple = (1,)
print(type(myTuple))
```

```python
myTuple = (1,2,3)    # tuple을 생성해요
myTuple = 1,2,3      # tuple을 생성해요 ()생략가능
# tuple은 list와 비슷해요!
print(myTuple[0])
print(myTuple[0:2])
# inedxing& slicing둘 다 가능
# myTuple[0] = 100     # 이거 안되요! read only
# tuple은 값을 바꿀 수 없어
```

```python
a,b,c = 0,20,30
(a,b,c) = 0,20,30
print(a)
```

```python
# list와 tuple
# list가 활용하기 좋아

a=[1,2,3] # list
b = tuple(a)
print(b)
# 파이썬의 가장 기본적인 구조 : list
# lsit 의 특별한 형태 : tuple
c = list(b)
print(c)
```

# Mapping Type - dic(dictionary)

```python
# Sequence type에는 결과적으로 3개가 존재
# list, tuple, range

# Mapping Type : dic(dictionary)
# key와 value의 쌍으로 표현되는 자료구조
# JSON ({name : "홍길동", age : 30 })
a = {"name" : "홍길동", "age" : 30}
# key값으로 value를 저장 (열쇠를 가지고 데이터를 저장)
print(type(a))
# key값은 불변의 값을 사용해야 해요!
# list는 key로 사용할 수가 없어요
# tuple은 key로 사용할 수 있어요
```

```python
a[10]="hohoho"
# 10 이 key , "hohoho" 가 value
print(a)
a["hobby"] = "Game"
print(a)
print(a["age"])
```

```python
b = {"name" : "홍길동", "age" : 30, "age" : 40}   # NG
print(b)
# key값은 unique해야해 이렇게 쓰면 안돼
```

```python
c = {"name" : "홍길동",("address",):"서울"}
# list는 키값으로 못씀
# tuple은 키값으로 쓸 수 있음
print(c)
print(c[("address",)])
```

```python
# dict의 함수
a = {"name" : "홍길동", "age" : 20, "address" : "서울"}
# value를 가져오고 싶어요 => key을 알아야해요
# a.keys()  # => dict_keys 자료구조로 리턴
            # list와 유사한 자료구조
# 키값을 다 뽑아서 list 만듦 (근데 진짜 list는 아님)
myKey = a.keys()
print(myKey)
# list는아닌데 list와 유사한거야
myList = list(myKey)
print(myList)
```

```python
a = {"name" : "홍길동", "age" : 20, "address" : "서울"}
myValue = a.values()   # dict_values 형태로 가져와요
print(myValue)
myItem = a.items()
print(myItem)
# key와 value를 쌍으로 tuple형태로
```

```python
# dict안에 있는 모든 key와 value를 출력하세요!
# for문을 이용해서
a = {"name" : "홍길동", "age" : 20, "address" : "서울"}
for k in a.keys():    # key에 대한 list 하나씩 꺼내와서 k에 mapping시킴
    print("{} - {}".format(k, a[k]))   # 문자열 포멧
    print("hohoho")
# 중괄호가 없어, 인덴트로 판단해야해

print("age" in a)  # 특정 key값이 dict에 잇는지 확인
```

```python
for k in a.keys():
    print("{} - {}".format(k, a[k]))
    print("hohoho")
    
for k in a.keys():
    print("{} - {}".format(k, a[k]))
    print("hohoho")
```

```python
a = {"name" : ["홍길동","최길동"], "age" : 20, "address" : "서울"}
# 복잡한 형식

print("age" in a, end="\n")
# 특정 key값이 dict에 있는지 확인

# value에 list, tuple, dict가 나옴
```

```python
# 19-11-13 python의 남은 built in type
## Set : type(set)
# set() 함수
## 어떻게 생겼나요? iteral 표현, 기호 => {}
## {} : 기본적으로 dict, dict는 {}안에 key와 value로 구성
a = { "name" : "홍길동", "age" : 20 }
b = { 1, 2, 3 }

print(type(a))
# set은 요소만 들어감
# set은 요소의 중복을 허용하지 않는 저장장소
# 기본적으로 순서가 정해지지 않음, 하나의 주머니!
# 문자열 입력시 한글자씩 인식되어 중복 제거됨
# set 자료구조는 중복을 허용하지 않는 저장장소
# 단, 순서가 있지 않음

## 간단하게 set예제를 들어보아요!
a = set([1,2,3,4,1,2,3,1,2,3]) # set은 중복을 배제해요!
print(a)
a = set("Hello")
print(a)  # 문자열을 set으로 만들면 어떻게 되나요?
```

```python
# set의 연산
a = {1,2,3,4} # set이 literal
b = {3,4,5,6}

# set => 집합 ( 집합연산, 교집합, 합집합, 차집합 )
print(a & b)   # & : 만약 set에 적용하면 교집합(intersection)
print(a.intersection(b))
print(a | b)  # | : 만약 set에 적용하면 합집합
print(a.union(b)) 
print(a - b)   # - : 만약 set에 적용하면 차집합(diffrerance)
print(a.difference(b))
```
