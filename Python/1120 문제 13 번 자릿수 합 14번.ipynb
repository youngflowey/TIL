```python
## 문제 13
# 다음과 같은 특성을 가지는 숫자의 개수를 구하는 프로그램을 작성해보자 
# 두개의 숫자(x,y)를 이용한다. 양의 정수 1과 100 
# x초과 y 미만 인 숫자중 각자리의 숫자를 모두 더한값이 5의 배수가 되는 숫자를 구하고 해당 숫자가 몇개인지출력 
# 1초과 100 미만 19개 
# 5초과 500 미만 98
# 100 초과 10000 미만은?
result = []
def prob(x):
    a= x%10
    b= x//10
    c= x//100
    d= x//1000
    e= x//10000
    f =x//100000
    if (a+b+c+d+e+f)%5==0:
        result.append(x)

for i in range(2,200):
    prob(i)
print(len(result))
print(result)
    
```

```python
%%time
def prove():
    Fresult=[]
    
    x,y = input().split()
    xi=int(x)
    yi=int(y)
    for num in range(xi+1,yi):
        result=[]
        num1=str(num)
        for i in range(len(num1)):
            result.append(int(num1[i]))
            
        if sum(result)%5==0:
            Fresult.append(num)
    return Fresult,len(Fresult)

prove()
```

```python
## 강사님 풀이 
def exercise13(x,y):
    nCount = 0
    
    for i in range(x+1,y):
        k=str(i)
        intSum=0
        for j in range(len(k)):
            intSum+=int(k[j])
        if intSum % 5 ==0:
            nCount+=1
    return nCount
exercise13(100,10000)
```

```python
## 14 번 6자리 이상 9자리 미만의 수를 하나 이용 
## 수의 중앙을 기준으로 두개의 수로 분리 
## 두개의 수로 분리한 다음 큰 수를 선택 
## 이 작업을 계속 진행  한자리수 나올때 까지 
def exer(a):
    a=str(a)
    if len(a) ==1 :
        return a 
    else :
        if len(a)%2==0:
            if a[:(len(a)//2)] >= a[len(a)//2:]:
                return exer(a[:((len(a)//2))])
            else:
                return exer(a[len(a)//2:])
        else:
            if (a[:len(a)//2]) >= (a[len(a)//2+1:]):
                return exer(a[:len(a)//2])
            else:
                return exer(a[len(a)//2+1:])
            
            
exer(6522345)
```
