```python
# Series를 생성해 보아요!
# NumPy array (ndarray) 와 비교해서 확인

# 먼저 ndarray부터 다시 한번 알아보아요!
arr = np.array([-1, 5, 8, 10])
        # 어떤 형태인지 list, tuple로 표현
print(arr.dtype)
arr = np.array([-1, 5, 8, 10], dtype=np.float64)
arr = np.array([-1, 5, 8, 10], dtype="float64")
        # dtype -> 타입 정해줄 수 있어
print(arr.dtype)

arr = np.array([-1, 5, 3.14, "Hello"])
# 이렇게 만들수 있어?
    # numpy array  -> 같은 데이터 타입이어야해
    # 그럼 못쓰는거아니야?
print(arr)
print(arr.dtype)

arr = np.array([-1, 5, 3.14, "Hello"], dtype = np.object)
# 객체형으로 잡음-> 각각의 객체로 인식 -> 객체기때문에 numpy array에 들어올 수 있음
# class로 부터 파생된 데이터 공간
# int, float, str이라는 class의 객체
print(arr)    # [-1 5 3.14 'Hello']
# 문자열이 아닌 그대로 나옴
print(arr.dtype)     # object
print(arr[2])        # 3.14
print(type(arr[2]))  # <class 'float'>
```

# Series

```python
## Series를 생성해보아요!

arr = np.array([-1,5,3.14,"Hello"], dtype = np.object)
s = pd.Series([-1,5,3.14,"Hello"], dtype = "object")
    # nbarray랑 비슷해
    # series내부에 nbarray가 들어있음
    # 외부에 덧씌워짐

print(s)
    # 출력내용 자체가 달라
    # 새로 뭘 붙여놓은것

print("=" * 30)

## 1. Series의 value부분 출력
# values : 시리즈가 가지고 있는 실제 데이터
print("Series의 value부분 : {}".format(s.values))
# [-1 5 3.14 'Hello'] -> ndarray
print("Series의 value부분의 type : {}".format(type(s.values)))
# <class 'numpy.ndarray'>
# numpy array를 기반으로 

## 2. Series의 data type(dtype)
print("Series의 value부분 dtype : {}".format(s.values.dtype))
print("Series의 value부분 dtype : {}".format(s.dtype))
# dtype : series안에 들어있는 데이터타입뭐야

## 3. Series의 index부분 출력
# series는 index 수정가능
print("Series의 index부분 : {}".format(s.index))
print("Series의 index부분 : {}".format(type(s.index)))
# rangeIndex
# index를 위한 판다스에서 새롭게 정의한 객체
# 데이터 / index 2가지로 구성되어있어 -> 따로 가능

# numpy array에 추가해서 새로운 자료구조
# 모아모아서 2차원 프레임을 만들어서 dataframe을 구성
```

```python
## Series는 index를 따로 지정할 수 있어요
## 숫자 index는 기본으로 제공
s = pd.Series([-1, 5, 8, 10],
             dtype = np.int32,
             index = ["a","c","k","f"] )
print(s)
# a    -1
# c     5
# k     8
# f    10

# 숫자 index대신 내가 정한 index를 사용가능

# index를 새롭게 지정하면 숫자 인덱스는 사용할 수 없나요?
print(s[1])   # 숫자 인덱스는 기본으로 사용 가능
print(s["k"])    # 새로 지정한 index를 이용해서 값을 access 가능

s = pd.Series([-1,5,8,10],
             dtype = np.int32,
             index = ["k","c","k","f"] )   # k인덱스 중복
print(s[1])  # 숫자 인덱스는 기본으로 사용가능
print(s["k"])
# 내가 새로 지정한 index라면 중복되도 사용가능하다
# 새로 지정한 index를 이용해서 값 access 가능
# 만약 찾으려는 index가 2개 이상 존재할 경우 그 결과가 Series로 return
```

```python
s = pd.Series([-1,5,8,10],
             dtype = np.int32,
             index = ["a","c","k","f"])
# ndarray에서 indexing 사용할 수 있고 slicing도 사용할 수 있어요
# Series도 slicing을 이용할 수 있어요!
print(s[0:2])   # Series가 return
print(s["a":"k"])    # 새로만든 index을 이용해서 slicing을 할때 범위에 조심!
                     # 둘다 include됨
    
print(s[s%2==0])    # Boolean indexing
# broadcasting -> 각각 나눠줌 -> 0이니? -> T/F -> mask

print(s[[0,2]])     # Fancy indexing
# 0번째 2번째 만

print(s.sum())
# series 안에있는거 다 더해
# ndarray의 집계함수 그대로이용가능
```

### for 문

```python
## for문
## for문을 list를 만들때 사용할 수 있어요!
myList = [1,2,3,4]
## 생성하고 싶은 list는 각 요소에 *2를 한 lsit
## -> [2,4,6,8]
result = myList * 2
print(result)
# -> [1, 2, 3, 4, 1, 2, 3, 4]
# list는 vector가 아니야
# *하면 더하기의 의미로 됨

# for문 돌리면될까?
result = list()
for i in myList:
    result.append(i*2)
print(result)
# [2, 4, 6, 8]

result = [ x*2 for x in myList]    # list 안에서 for문이 돌아
# 총 요소 4개 만들어지고 각각의 요소
print(result)

result = [ x*2 for x in myList if x > 2]
print(result)
# 기존 데이터를 반복하고 조건잡아서

# dict에도 사용할 수 있나요??
## ==> ooo
result = {}  # dict아니면 set

result = { "stu" + str(x) : x**2  for x in range(0,10)}
print(result)
```

### Delta

```python
from datetime import date, datetime, timedelta

today = datetime.today()
print(today)
day = timedelta(days = -1)    # -1 : 하루전, 2: 이틀 후
# > 2019-11-20 11:16:24.274638

# 날짜에 대한 델타를 구할거야
# 뒤에 델타 값을 입력해야해

result = today + day  # 하루 전 날짜
print(result)
# > 2019-11-19 11:16:24.274638

# delta <- weeks, days, hours, minutes, seconds
#          years, months (x) -> 윤년 윤달과 관련있기 때문에
day = timedelta(hours = -1)
result = today + day
print(result)
# > 2019-11-20 10:16:24.274638

from dateutil.relativedelta import relativedelta

today = datetime.today()
print(today)
# > 2019-11-20 11:16:24.274638

day = relativedelta(months = -1)
# years, months
result = today + day
print(result)
# > 2019-10-20 11:16:24.274638

# 오늘날짜가 3월 31일 => 1달전? 2월 28일, 2월 29일

from dateutil.parser import parse

today = parse("2019-03-31")
print("오늘의 날짜 : {}".format(today))
# > 오늘의 날짜 : 2019-03-31 00:00:00
day = relativedelta(months=-1)
print(today+day)
# > 2019-02-28 00:00:00
```

```python
import numpy as np
import pandas as pd

data = {
    "이름" : ["스폰지밥","뚱이","징징이","다람이"],
    "학과" : ["요리","바보","클라리넷","과학"],
    "학년" : [1,3,4,2],
    "학점" : [3.1, 2.1, 3.8, 4.5]
}

df = pd.DataFrame(data,
                 columns = ["학과","학년","이름","학점","등급"],       
                 index = ["one","two","three","four"])   
display(df)

# Boolean indexing은 mask를 사용해요!

# 학점이 3.0 이상인 것 뽑고싶어
df['학점'] >= 3.0     # [True, False, True, True]

# 학점이 3.0 이상인 사람의 학과와 이름을 출력하세요!
# df.loc[행,열]
# 행 부분에 mask 로 들어감

df.loc[df['학점'] >= 3.0,["학과","이름"]]

```

```python
## 분석용 함수(통계함수)
## 평균, 편차, 분산, 표준편차, 공분산, 상관계수

## 평균(mean)

## 편차(deviation) : 확률변수 X와 mean의 차이
# X => 8 5 7 3 2
# mean => 5
# deviation => 3 0 2 -2 -3 => sum은 0
# 분산(variance) : 데이터의 흩어짐 정도를 알기 위해서 사용하는 값으로 편차의 제곱의 평균
# 표준편차(standard deviation) : 분산의 제곱근
                               # 분산의 단위 문제를 해결하고자 사용

import numpy as np
    
arr = np.array([4,6,1,3,8,8], dtype=np.int32)

print(arr)
print(arr.sum())
print(np.sum(arr))  # 합 .  안으로 들어오는인자 싹다 sum
print(arr.mean())   # 평균
print(arr.var())    # 분산
print(arr.std())    # 표준편차

# 공분산
# 데이터가 변하는 경향 알 수 있음
# 데이터의 강도 는 모름 (관련있는 변수냐 아니냐 는 파악할 수 없어)
# 공분산 : 두개의 확률 변수 X,Y가 있을 때
#          이 두개의 확률변수의 관계를 보여줄 때 사용해요!
# 공분산 : 두 확률변수의 편차의 곱의 평균으로 표현
#          데이터가 평균으로부터 얼마나 떨어져있느냐를 보여줌
# 그래프를 그려보아요!
import numpy as np
import matplotlib.pyplot as plt

# 만약 module이 없다고 error가 나오면
# conda install matplotlib

np.random.seed(4)     # 항상 같은 난수가 발생하도록 seed 처리

x = np.random.randint(-20,20,(10,))    # -20부터 19까지 , 1차원으로 10개
print(x)
y = np.random.randint(-10,10,(10,))  
# 두개 확률 변수 난수의 형태로

x_mean = x.mean()
y_mean = y.mean()

x_deviation = x-x_mean
y_deviation = y-y_mean

print(x_deviation)
print(y_deviation)

result=0
for i in range(10):
    result += (x_deviation[i] * y_deviation[i])
# 편차의 곱의 합
# 우리가 원하는 값은 편차의 곱의 합의 평균
myCov = result/9    # 모공분산 : 원래 공식
                      # 표준공분산 : n-1  -> 추정치가 좋아짐 
print(myCov)
print(np.cov(x,y))



## scatter(산점도)

plt.scatter(x,y,color="red")

# x가 얼마나 변함에따라 y가 얼마나 변하는지

plt.scatter(x_mean,y_mean,color="blue")   # 평균점은 어디
plt.show()

# 공분산이 양수 -> 하나가 증가하면 같이 증가
#          음수 -> 하나가 증가하면 하나는 감소
# 단위 때문에 제각각 -> 강도를 표현하기 의미 X
# 연결성을 알 순 없어, 경향만을 파악
# 공분산으로 상관계수 -> 두 변수의 연결성

# 공분산 : 두확률변수의 관계(경향)을 의미
        # 상관 강도는 알 수 없어
          
```

```python
## 상관관계 (correlation)
# 두 대상이 서로 연관성이 있다고 추측되는 관계
## 상관계수 (correlation coefficient) : 일반적으로 피어슨 상관계수를 이용해요(다른 상관계수도 존재.)
## 공분산을 각 확률변수의 표준편차의 곱으로 나눠준값
## -1 ~ 1 사이의 값으로 표현되요!
# 두 개가 독립이면 공분산은 0이다
# 공분산이 0이지만 두개가 독립이지 않을 수도 있다
# 양수일 경우는 양의 상관관계라고 표현해요!
# 음수일 경우는 부적 상관관계라고 표현해요!
## 성적 & 자존감
## 온라인 게임 & 폭력성 ( 폭력성을 수치화해서 ) -> 연관성 있다
# 상관계수(상관관계)는 인과관계로 설명하지 않는다 -> 오류발생함
# 온라인 게임많이 하면 폭력적이야! -> 프로게이머들 조폭이야? 아니야
# 인과관계는 회귀분석(regression)을 통해서 알 수 있어요! 

import numpy as np
import pandas as pd
import json

# json 파일을 불러들이려면 열어야해
file_KOSPI = open("C:/python_DA/data/주식데이터/KOSPI.json",'r')
# 파일 열기 : open()
# df_KOSPI = pd.DataFrame(json.load(file_KOSPI))
# display(df_KOSPI.head())
# df_KOSPI["Close"]
series_KOSPI = pd.DataFrame(json.load(file_KOSPI))["Close"]

file_SE = open("C:/python_DA/data/주식데이터/SE.json",'r')
series_SE = pd.DataFrame(json.load(file_SE))["Close"]

## 상관계수를 구하기 위해서 일단 공분산을 구한 후 각 표준편차의 곱으로 나눠줘야 해요!
# 넘파이가 가지고 있는 함수를 이용
print(np.corrcoef(series_KOSPI,series_SE))
# corrcoef : 피어슨 상관계수
# 두개의 시리즈를 해당 함수의 인자로 넣어주면,
# 0.9 -> 굉장히 가까움
# 두개 버리지 않고 이용해야 겠군!

file_PUSAN = open("C:/python_DA/data/주식데이터/부산산업.json",'r')
series_PUSAN = pd.DataFrame(json.load(file_PUSAN))["Close"]

## 3개의 Series를 DataFrame으로 만들어보아요!
data = {
    "KOSPI" : series_KOSPI,   # series를 가지고 dict -> dataframe
    "SE" : series_SE,
    "PUSAN" : series_PUSAN
}
df = pd.DataFrame(data)
display(df.head())
display(df.corr())
# dataframe은 corr까지 쓴다
# 여러개의 col에 대해서 한방에 쉽게 계산 가능
display(df.cov())
# 음의 관계 양의 관계만 표현, 강도는 표현 X
# 별로 사용은 X
```

```python
## DataFrame에서 sum함수와 mean함수의 사용
import numpy as np
import pandas as pd

## 중첩리스트를 이용해서 DataFrame을 생성해 보아요!
data = [[2, np.nan],
       [7 , -3],
       [np.nan, np.nan],
       [1, -2]]

df = pd.DataFrame(data,
                 columns = ["one","two"],
                 index = ['a','b','c','d'])
display(df)
print(df.sum())
# 안에 있는거 싹다 다 더하는거
# Dataframe에서는 sum()함수에 대해 aixis를 주지 않으면 axis=0으로 세팅
# series로 떨어뜨림
# nan을 0으로 간주 -> 없는걸로 간주

display(df.sum(axis=1))     # 열방향
display(df["two"].sum())
display(df.loc["b"].sum())
# col 가져올때는 indexing
# row 가져올때는 loc indexing  ( 숫자 못써 )

display(df.mean(axis=0, skipna=False))    # skipna : na를 skip할거니
                                  # NaN은 실수처리, 모든연산결과는 NaN
# 결측값을 날리면 위험해 -> 실데이터도 날라가게 됨 -> 대치해서 사용하는게 가장 좋은방법
display(df.mean(axis=0, skipna=True)) 

# 'one' column의 결측값은 'one' column의 평균으로
# 'two' column의 결측값은 'two' column의 최소값으로 대체해 보아요!
one_avg = df["one"].mean(axis = 0, skipna=True)
print(one_avg)
two_min = df["two"].min(axis=0)

df["one"] = df["one"].fillna(value=one_avg)
# 원래의 값에 대치
# 데이터프레임을 조작 -> 원본이 바뀌지 않아
# 처리된 결과를 새로운 데이터프레임으로 받아서 사용

df["two"] = df["two"].fillna(value=two_min)

display(df)
```

```python
# inner join, outer join(full), left join, right join
import numpy as np
import pandas as pd

data1 = {
    '학번' : [1,2,3,4],
    '이름' : ['스폰지밥','뚱이','징징이','다람이'],
    '학년' : [2,3,3,1]
}

data2 = {
    '학번' : [1,2,4,5],
    '학과' : ['컴퓨터','미술','철학','사회'],
    '학점' : [1.3,3.5,4.3,2.3]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
display(df1)
display(df2)

display(pd.merge(df1, df2, on = '학번', how = 'inner'))
# pandas가 가지고 있는 merge기능
# 무슨기준, 어떤 방식?
# on = '결합할 컬럼명'
# how = '결합 방식'
```

### Grouping

```python
import numpy as np
import pandas as pd

df = pd.DataFrame({'이름' : ['스폰지밥','뚱이','다람이','집게사장','징징이'],
                   '성적' : [1.5,4.4,4.5,2.3,4.0],
                   '학과' : ['컴퓨터','철학','컴퓨터','철학','컴퓨터'],
                   '학년' : [1,2,3,2,2]})
display(df)
```

```python
# 학과를 기준으로 grouping
# grouping 대상 : '성적'

dept = df['성적'].groupby(df['학과'])
# 결과는 2개 떨어지고
# 컴퓨터와 철학으로 나눠짐
# 학과별로 성적에 대한 평균
dept.mean()
```

```python
# 학과와 학년을 기준으로 성적을 grouping
# 총그룹이 4개
# 학과와 학년을 이용한 grouping

dept = df['성적'].groupby([df['학과'],df['학년']])
dept.mean()

# 2차 인덱스 / multi index로 빠짐
```

```python
dept.mean().unstack()
```
